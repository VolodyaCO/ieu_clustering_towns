```python
import pandas as pd
from geopy.geocoders import Nominatim
import re

geolocator = Nominatim(user_agent='myapplication', timeout=3)
```

```python
df = pd.read_excel('Areas y población Mun_COL 2005_2017.xlsx').iloc[:-1,:]
```

```python
df.head()
```

```python
towns = df['MPIO'].values
state = df['DPNOM'].values
```

```python
locations = []
for town, st in zip(towns, state):
    town = re.sub(r'\([^)]*\)', '', town)
    query = "{0}, {1}, Colombia".format(town, st)
    print(query)
    location = geolocator.geocode(query)
    if not location:
        location = geolocator.geocode("{0}, Colombia".format(town))
    print(location)
    try:
        locations.append((town, st, location.latitude, location.longitude))
    except Exception as e:
        print(e)
```
